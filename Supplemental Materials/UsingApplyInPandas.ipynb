```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings("ignore")
```

```python
penguins_df = pd.read_csv("https://raw.githubusercontent.com/mcnakhaee/palmerpenguins/master/palmerpenguins/data/penguins-raw.csv")
penguins_df.dropna(inplace = True)
penguins_df.reset_index(inplace = True, drop = True)
penguins_df.head(3)
```

```python
# Can we switch this column from Female, Male --> 1, 0 

#Pandas Apply

#1. Create the necessary function

def swapSex(value):
    
    if value == "FEMALE":
        
        return 1
    
    return 0

penguins_df["Sex"] = penguins_df["Sex"].apply(swapSex)
```
