```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
# import else here'
```

```python
df = pd.read_csv("Data/milk.csv")
```

### Description of Histograms

This plot shows that the grade of milk based on the turbidity value is actually ...


```python
hist_turbidity = px.histogram(df, x = "Grade", color = "Turbidity", width = 800, title = "Grade of Milk: Turbidity", color_discrete_sequence=['#7995d0', '#9c948c'])
hist_turbidity.show()
hist_odor = px.histogram(df, x = "Grade", color = "Odor", width = 800, title = "Grade of Milk: Odor", color_discrete_sequence=['#7995d0', '#9c948c'])
hist_odor.show()
hist_taste = px.histogram(df, x = "Grade", color = "Taste", width = 800, title = "Grade of Milk: Taste", color_discrete_sequence=['#7995d0', '#9c948c'])
hist_taste.show()
hist_fat = px.histogram(df, x = "Grade", color = "Fat ", width = 800, title = "Grade of Milk: Fat", color_discrete_sequence=['#7995d0', '#9c948c'])
hist_fat.show()
```

```python
px.scatter(df, x='Temprature', y='Grade', color='Fat ')

#Shows that high grade milk has lower temprature and more fat 
```
