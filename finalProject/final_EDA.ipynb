```python
hist_turbidity = px.histogram(df, x = "Grade", color = "Turbidity", width = 800, title = "Grade of Milk: Turbidity", color_discrete_sequence=['#7995d0', '#9c948c'])
hist_turbidity.show()
hist_odor = px.histogram(df, x = "Grade", color = "Odor", width = 800, title = "Grade of Milk: Odor", color_discrete_sequence=['#7995d0', '#9c948c'])
hist_odor.show()
hist_odor.write_html("odor.html")
hist_taste = px.histogram(df, x = "Grade", color = "Taste", width = 800, title = "Grade of Milk: Taste", color_discrete_sequence=['#7995d0', '#9c948c'])
hist_taste.show()
hist_fat = px.histogram(df, x = "Grade", color = "Fat ", width = 800, title = "Grade of Milk: Fat", color_discrete_sequence=['#7995d0', '#9c948c'])
hist_fat.show()
```
