```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
# import else here'
```

```python
df = pd.read_csv("Data/milk.csv")
```

### Description of Histograms

This plot shows that the grade of milk based on the turbidity value is actually ...


```python
hist_turbidity = px.histogram(df, x = "Grade", color = "Turbidity", width = 800, title = "Grade of Milk: Turbidity")
hist_turbidity.show()
hist_odor = px.histogram(df, x = "Grade", color = "Odor", width = 800, title = "Grade of Milk: Odor")
hist_odor.show()
hist_taste = px.histogram(df, x = "Grade", color = "Taste", width = 800, title = "Grade of Milk: Taste")
hist_taste.show()
hist_fat = px.histogram(df, x = "Grade", color = "Fat ", width = 800, title = "Grade of Milk: Fat")
hist_fat.show()
```

```python
px.scatter(df, x='Temprature', y='Grade', color='Fat ')

#Shows that high grade milk has lower temprature and more fat 
```

```python
fig3 = px.pie(df, values='pH',names= 'pH', title='Milk Grade')
fig3.show()

#shows the distrubition of milk pH
```
