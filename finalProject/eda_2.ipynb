```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv("Data/milk.csv")
```

```python
# Relationship between pH and temperature, labeled with grade
scatter_pH_temp = px.scatter(df, x = "pH", y = "Temprature", color = "Grade", width = 800, title = "Grade of Milk: pH vs Temperature")
scatter_pH_temp.show()
```

```python
# Histograms representing order and turbidity labeled by grade
hist_turbidity = px.histogram(df, x = "Grade", color = "Turbidity", width = 800, title = "Grade of Milk: Turbidity")
hist_turbidity.show()
hist_odor = px.histogram(df, x = "Grade", color = "Odor", width = 800, title = "Grade of Milk: Odor")
hist_odor.show()
hist_taste = px.histogram(df, x = "Grade", color = "Taste", width = 800, title = "Grade of Milk: Taste")
hist_taste.show()
hist_fat = px.histogram(df, x = "Grade", color = "Fat ", width = 800, title = "Grade of Milk: Fat")
hist_fat.show()
```

```python
scatter_color = px.scatter(df, x = "pH", y = "Colour", color = "Grade", width = 800, title = "Grade of Milk: pH vs Color")
scatter_color.show()
```
